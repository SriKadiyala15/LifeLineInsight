```python
#Load & Inspect the Dataimport pandas as pd
df = pd.read_csv("C:/Users/ADMIN/Desktop/Imp/Projects/LifelineInsight/Suicide_Dataset.csv")
df.head()
```

```python
#Standardizing column names
df.columns = ['State', 'Year', 'Type_Code', 'Type', 'Gender', 'Age_Group', 'Total']
```

```python
#Checking for nulls & data types
df.isnull().sum()
df.dtypes
```

```python
#Convert Year to int
df['Year'] = df['Year'].astype(int)
```

##Exploratory Data Analysis

```python
#Total Suicides per Year
df.groupby('Year')['Total'].sum().plot(kind='line', title='Suicides Over Years')
```

```python
#Total Suicides by Gender
df.groupby('Gender')['Total'].sum().plot(kind='bar', title='Suicides by Gender')
```

```python
#Top 10 States by Total Suicides
df.groupby('State')['Total'].sum().sort_values(ascending=False).head(10).plot(kind='barh')
```

```python
#Suicides by Age Group
df.groupby('Age_Group')['Total'].sum().plot(kind='bar')
```

```python
#Suicides by Cause
df.groupby('Type')['Total'].sum().sort_values(ascending=False).head(10).plot(kind='barh')
```

```python
#Saving Final Grouped Data for Tableau
df.to_csv("suicide_analysis_final.csv", index=False)
```
